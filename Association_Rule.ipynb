# **Association Rule Mining**

Association rule mining is a technique in data mining for discovering interesting relationships (associations) between variables in large datasets. It is widely used in market basket analysis, where the goal is to find rules that predict the occurrence of an item based on the occurrences of other items.

Apriori Algorithm
The Apriori algorithm is a classical algorithm for mining frequent itemsets and generating association rules. It operates on the principle that any subset of a frequent itemset must also be frequent. The algorithm proceeds in two main steps:

1) Frequent Itemset Generation:

Candidate Generation: Generate candidate itemsets of length k from frequent itemsets of length k-1.

Support Counting: Calculate the support of these candidates and prune the ones that do not meet the minimum support threshold.

2) Rule Generation:

Generate rules from the frequent itemsets that meet the minimum confidence threshold.

# **Key Metrics in Apriori and Association Rule Mining**


Interpretation

# a) Lift:

Lift measures how much more likely two items are to be bought together than if they were independent. A lift value greater than 1 indicates a positive association.
The pairs with the highest lift values represent the strongest associations. For instance, if "Bread" and "Butter" have a high lift value, it means they are frequently bought together significantly more than would be expected by chance.

Example:

High Lift Values (Strong Associations) Bread -> Eggs and Eggs -> Bread:

Lift: 10.0 Interpretation: Bread and Eggs have a very strong association. If a customer buys Bread, they are 10 times more likely to also buy Eggs compared to buying Eggs independently.

# b) Confidence:

Confidence measures the likelihood that item B is bought when item A is bought. It is a measure of the rule's strength.
High confidence values indicate that if the base item is bought, the dependent item is also likely to be bought. For example, a high confidence value for the pair "Bread" -> "Butter" suggests that customers who buy bread often buy butter.

Example:


High Confidence Values (Strong Rules) Bread -> Butter

Confidence: 1.0 Interpretation: Every time Bread is bought, Butter is also bought. This is a very strong rule with 100% confidence. Bread -> Eggs and Eggs -> Bread

# c) Support:

Support measures the frequency of the itemset appearing in the dataset. It is the proportion of transactions that contain the itemset.
High support values indicate that the itemset is common in the dataset. For example, if "Milk" and "Bread" have high support, it means they are commonly bought together across many transactions.

Example:


High Support Values (Common Item Pairs) Juice -> Sugar and Sugar -> Juice

Support: 0.10 Interpretation: 10% of all transactions include both Juice and Sugar. These item pairs are quite common in the dataset. Juice -> Cheese and Cheese -> Juice


By analyzing these metrics, businesses can identify strong association rules to make strategic decisions, such as placing frequently bought items together in the store or offering promotions for commonly associated products.

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
# Loading the data
data = pd.read_csv("/content/Market_Basket_Optimisation.csv")
DATA = data.copy()
```

```python
# Preprocessing the data to ensure all items are strings and remove any None or NaN values
transactions = [[str(item) for item in transaction if item is not None] for transaction in data]
```

```python
# Applying the Apriori algorithm
association_rules = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
association_results = list(association_rules)
```

```python
# Extracting the rules and their metrics
confidence = []
lift = []
support = []
base_item = []
dependent_item = []

for rule in association_results:
    for item in rule.ordered_statistics:
        if len(item.items_base) > 0 and len(item.items_add) > 0:
            lift.append(item.lift)
            confidence.append(item.confidence)
            support.append(rule.support)
            base_item.append(list(item.items_base)[0])
            dependent_item.append(list(item.items_add)[0])
```

```python
# Creating a DataFrame with the results
df = pd.DataFrame({
    "Base_Item": base_item,
    "Dependent_Item": dependent_item,
    "Support": support,
    "Confidence": confidence,
    "Lift": lift
})
```

```python
# Sorting and visualizing the results
sorted_by_lift = df.sort_values(by="Lift", ascending=False, ignore_index=True)
sorted_by_confidence = df.sort_values(by="Confidence", ascending=False, ignore_index=True)
sorted_by_support = df.sort_values(by="Support", ascending=False, ignore_index=True)
```
